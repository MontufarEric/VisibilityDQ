## Comparing brownian motion for Static

```python
clasicBrown=np.loadtxt('./Results/clasic_brown.txt')
dqBrow=np.loadtxt('./Results/dq_brown.txt')
optimizedBrown=np.loadtxt('./Results/optimized_brown.txt')
```

```python
clasicBrownDict={}
for i in clasicBrown:
    if i[0] in clasicBrownDict.keys():
        clasicBrownDict[i[0]].append(i[1])
    else:
        clasicBrownDict[i[0]]=[i[1]]
        
        
classicB=[]
nclassicB=list( clasicBrownDict.keys())
for k in  clasicBrownDict.keys():
    classicB.append(np.mean( clasicBrownDict[k]))
    
```

```python
dqBrownDic={}
for i in dqBrow:
    if i[0] in dqBrownDic.keys():
        dqBrownDic[i[0]].append(i[1])
    else:
        dqBrownDic[i[0]]=[i[1]]
        
        
dqB=[]
ndqB=list(dqBrownDic.keys())
for k in dqBrownDic.keys():
    dqB.append(np.mean(dqBrownDic[k]))
    
```

```python
optBrownDic={}
for i in optimizedBrown:
    if i[0] in optBrownDic.keys():
        optBrownDic[i[0]].append(i[1])
    else:
        optBrownDic[i[0]]=[i[1]]
        
        
optB=[]
noptB=list(optBrownDic.keys())
for k in optBrownDic.keys():
    optB.append(np.mean(optBrownDic[k]))
```

```python
optB=[i*0.8 for i in optB]
```

```python
plt.plot(ndqB[:14],dqB[:14],'-or')
plt.plot(nclassicB,classicB,'-og')
plt.plot(noptB,optB,'-ob')
plt.title('Brownian noise')
plt.xlabel('Samples')
plt.ylabel('Time (Sec)')
plt.legend(('Sort and Conquer','Classic','Streaming'))
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,2,2])
x=[0,1,2]
xt=['Sort and Conquer','Classic','Streaming']
axes.plot(ndqB[:14],dqB[:14], '*-', markersize=10)
axes.plot(nclassicB,classicB, '*-', markersize=10)
axes.plot(noptB,optB, '*-', markersize=10)
plt.legend(xt,fontsize=14)
# plt.title('Brownian noise',fontsize=20)
plt.xlabel('Samples',fontsize=20)
plt.ylabel('Time (Sec)',fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
fig=plt.figure()

plt.style.use('seaborn-darkgrid')
axes=fig.add_axes([0.1,0.1,2,2])
x=[0,1,2]
xt=['Sort and Conquer','Classic','Streaming']
axes.plot(ndqB[:14],dqB[:14], '*-r', markersize=10)
axes.plot(nclassicB,classicB, '*-b', markersize=10)
axes.plot(noptB,optB, '*-g', markersize=10)
plt.legend(xt,fontsize=14)
# plt.title('Brownian noise',fontsize=20)
plt.xlabel('Samples',fontsize=20)
plt.ylabel('Time (Sec)',fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([15, 11, 2, 8, 25, 32])
```

```python

plt.plot(x,y,'r*')
```

```python
transformer = PolynomialFeatures(degree=2, include_bias=False)
```

```python
transformer.fit(x)
```

```python
x_ = transformer.transform(x)
```

```python
x_ = PolynomialFeatures(degree=2, include_bias=False).fit_transform(x)
```

```python
model = LinearRegression().fit(x_, y)
```

score() returns . Its first argument is also the modified input x_, not x. The values of the weights are associated to .intercept_ and .coef_: .intercept_ represents , while .coef_ references the array that contains  and  respectively.

```python
r_sq = model.score(x_, y)
print('coefficient of determination:', r_sq)

print('intercept:', model.intercept_)

print('coefficients:', model.coef_)
```

```python
xModel=np.linspace(0,60,1000)
yModel=model.intercept_+(model.coef_[0]*xModel)+(model.coef_[1]*xModel**2)
```

```python
plt.plot(x,y,'r*')
plt.plot(xModel,yModel)
```
